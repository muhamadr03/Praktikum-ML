```python
import pandas as pd
# membaca file day.csv

try:
  data_asli = pd.read_csv(path + '/day.csv')
  print("Berhasil membaca file 'day.csv'.")
  print(f"Total ada {len(data_asli)} baris data.\n")
except FileNotFoundError:
  print("Error: Pastikan file 'day.csv' ada")
  exit()
```

```python
# Mengacak Data
data_acak = data_asli.sample(frac=1, random_state=42).reset_index(drop=True)
print("Semua baris data sudah berhasil diacak.\n")
```

```python
# Menghitung Data
jml_total_data = len(data_acak)

# Data testing 20% dari total data
jml_data_testing = int(0.2 * jml_total_data)

# Sisa data untuk training dan testing 80% dari total data
jml_training_testing = jml_total_data - jml_data_testing

# Jumlah data validation 10% dari sisa data
jml_data_validation = int (0.1 * jml_training_testing)

# Jumlah data training sisa dari semua data
jml_data_training = jml_training_testing - jml_data_validation

print(f"Jumlah data training adalah: {jml_data_training}")
print(f"Jumlah data validation adalah: {jml_data_validation}")
print(f"Jumlah data testing adalah: {jml_data_testing}")
print(f"Total seluruh data adalah: {jml_data_training + jml_data_validation + jml_data_testing}")
```

```python
# Menampilkan 5 baris pertama data training
display(data_acak[:jml_data_training].head())

# Menampilkan 5 baris pertama data validation
print("\n========================================")
print("         (b) DATA VALIDATION")
print("========================================")
print("Lima baris pertama data validation:")
display(data_acak[jml_data_training:jml_data_training+jml_data_validation].head())

# Menampilkan 5 baris pertama data testing
print("\n========================================")
print("          (c) DATA TESTING")
print("========================================")
print("Lima baris pertama data testing:")
display(data_acak[jml_data_training+jml_data_validation:].head())
```
