Pemilihan Fitur

```python
# Fitur untuk clustering
X = df.drop(["Id","Species",], axis=1)
X.head()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 10))

# Histogram 1: Sepal Length
axes[0,0].set_title("Sepal Length")
axes[0,0].hist(X['SepalLengthCm'], bins=5)

# Histogram 2: Sepal Length
axes[1,0].set_title("Petal Length")
axes[1,0].hist(X['PetalLengthCm'], bins=5)

# Histogram 3: Sepal Length
axes[0,1].set_title("Sepal Width")
axes[0,1].hist(X['SepalWidthCm'], bins=5)

# Histogram 4: Sepal Length
axes[1,1].set_title("Petal Width")
axes[1,1].hist(X['PetalWidthCm'], bins=5)

plt.tight_layout()
plt.show()
```

Normalisasi data

```python
scaler = StandardScaler()
X = scaler.fit_transform(X)
X[:5]
```

Mencari nilai K terbaik

```python
inertia_list = []
K_range = range(1,10)

for k in K_range:
  km = KMeans(n_clusters=k, random_state=42)
  km.fit(X)
  inertia_list.append(km.inertia_)

plt.figure(figsize=(7,5))
plt.plot(K_range, inertia_list, marker='o')
plt.title('Elbow Method')
plt.xlabel('Jumlah cluster (K)')
plt.ylabel('Inertia')
plt.grid(True)
plt.show()
```

```python
kmeansh = KMeans(n_clusters=3,
                 init="k-means++",
                 random_state=42)
```

```python
# Latih Model
kmeansh.fit(X)
```

```python
# Prediksi Label untuk data X
labels = kmeansh.fit_predict(X)
```

```python
df['Cluster'] = labels
```

```python
print("Cluster ditemukan:", df['Cluster'].unique())
```

```python
# silhouette_score
sll_score = silhouette_score(X, labels)
print("Silhouette Score:", sll_score)
```

```python
from sklearn.metrics import confusion_matrix, classification_report

# Encode the 'Species' column into numerical labels
le = LabelEncoder()
df['SpeciesEncoded'] = le.fit_transform(df['Species'])

print("\n ===== CONFUSION MATRIX ===== ")
cm = confusion_matrix(df["SpeciesEncoded"], df["Cluster"])
print(cm)
```

```python
plt.figure(figsize=(6,5))
sns.heatmap(
    cm, annot=True, fmt='d', cmap='Blues',
    xticklabels=["Cluster 0", "Cluster 1", "Cluster 2"],
    yticklabels=le.classes_
)
plt.title("Confusion Matrix: K-Means vs Label Asli")
plt.xlabel("Cluster Prediksi")
plt.ylabel("Label Asli")
plt.show()
```

```python
#  2D dan 3D Visualisasi Hasil Model K-Means
plt.figure(figsize=(7,5))
sns.scatterplot(
    x=df["SepalLengthCm"],
    y=df["SepalWidthCm"],
    hue=df["Cluster"],
    palette="deep"
)
plt.title("Visualisasi K-Means Clustering (2D)")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.show()
```

```python
#Visualising the clusters
plt.scatter(X[df['Cluster'] == 0, 0], X[df['Cluster'] == 0, 1], s = 100, c = 'red', label = 'Iris-setosa')
plt.scatter(X[df['Cluster'] == 1, 0], X[df['Cluster'] == 1, 1], s = 100, c = 'blue', label = 'Iris-versicolour')
plt.scatter(X[df['Cluster'] == 2, 0], X[df['Cluster'] == 2, 1], s = 100, c = 'green', label = 'Iris-virginica')

#Plotting the centroids of the clusters
plt.scatter(kmeansh.cluster_centers_[:, 0], kmeansh.cluster_centers_[:,1], s = 100, c = 'yellow', label = 'Centroids')

plt.legend()
```
